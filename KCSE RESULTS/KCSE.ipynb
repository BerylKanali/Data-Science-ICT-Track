### Importing libraries and the dataset

```python
##importing libaries 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
##loading the dataset
kcse=pd.read_csv('KCSE_2016.csv')
```

### Data Preperation
Preparing the data to be in a form that is clear and easy to use.

```python
##viewing the first five rows of the data
kcse.head(5)
```

```python
#spliting the Gender column into two columns, gender and year
#I want to have two categorical columns, year and gender to use them for visualisation
kcse[['gender', 'Year']]=kcse.Gender.str.split("(", expand=True,)
kcse
```

```python
#stripping the closing bracket from the year column
kcse['Year']=kcse["Year"].str.strip(')')
kcse
```

```python
#Creating a column called 'Total' which is a sum for all the numeric values in the rows
#This will be a colum of all the students who sat for the exams
kcse['Total']=kcse.sum(axis=1)
kcse.head(5)
```

```python
#Dropping the Gender column since I have split it into two columns andI do not need it anymore
kcse=kcse.drop(['Gender'], axis = 1)
```

```python
#How my data looks like after dropping Gender column
kcse.head(5)
```

```python
#making my data clear to read and understand by grouping it by categories available that is Year and Gender
#It will make it easier to know values per year given a certain gender
kcse.groupby(["Year","gender"]).sum().reset_index()
kcse.head(18)
```

```python
#Rearranfing column order
kcse=kcse[["Year","gender","A","A-","B+","B","B-","C+","C","C-","D+","D","D-","E"]]
kcse.head(5)
```

```python
#Creating a new dataframe from the kcse dataframe
#This dataframe features grades up to C+ which is required for direct entry to university. I will useit to conduct further analysis
kcse1=kcse[kcse.columns[0:8]]
kcse1.head(5)
```

```python
#Adding a column to feature the sumof candidates with C+ and above
kcse1['Total1']=kcse1.sum(axis=1)
kcse1.head(5)
```

```python
#Joining the column "Total" from the kcse dataframe to the new dataframe
kcse2=kcse1.join(kcse[["Total"]])
kcse2
```

```python
kcse1["Percentage"]=kcse1["Total1"]/kcse["Total"]*100
kcse1
```

### Explatory Data Analysis

```python
#Graph showing number of candidates who sat for exams over the years
sns.set_style("darkgrid")
sns.catplot(x='Year',y='Total',hue="gender",kind='bar',hue_order=["ALL"],data=kcse,height=6,aspect=2)
plt.xlabel("Year",fontsize=15)#labelling the x axis
plt.ylabel("Total number of candidates",fontsize=15)#labelling the y axis
plt.title("Total Number Candidates Per Year",fontsize=20)#Title of the plot
plt.savefig("figure1.png",bbox_inches="tight")
```

-As the years progess more students sit for the national exams. Meaning more kenyans are getting an education.

```python
#Graph showing the distributio of total number of candidates over the years by gender
sns.set_style("darkgrid")
sns.catplot(x='Year',y='Total',hue="gender",kind='bar',hue_order=["FEMALE","MALE"],data=kcse,height=6,aspect=2)
plt.xlabel("Year",fontsize=15)#labelling the x axis
plt.ylabel("Total number of candidates",fontsize=15)#labelling the y axis
plt.title("Total Candidates Per Year by Gender",fontsize=20)#Title of the plot
plt.savefig("figure2.png",bbox_inches="tight")
```

-From the gragh, more males sit for the national exams as compaed to females with an exception of 2016
